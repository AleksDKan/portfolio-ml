Since there are only 4 in 'Other' category from 'transmission' and 5 cars in 'Electric' from 'fuelType', let's add these to another categories: 

```python
data_cleaned['transmission'] = data_cleaned['transmission'].replace('Other', 'Manual')
data_cleaned['fuelType'] = data_cleaned['fuelType'].replace('Electric', 'Manual')
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['year'],data_cleaned['price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['mileage'],data_cleaned['price'])
ax2.set_title('Price and Mileage')
ax3.scatter(data_cleaned['mpg'],data_cleaned['price'])
ax3.set_title('Price and Mpg')
ax4.scatter(data_cleaned['engineSize'],data_cleaned['price'])
ax4.set_title('Price and EngineSize');
```

From the subplots and the probability distribution function of price, we can easily determine that 'Price' is exponentially distributed
A good transformation in that case is a log transformation

```python
log_price = np.log(data_cleaned['price'])
data_cleaned['log_price'] = log_price
data_cleaned
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True, figsize =(15,3)) #sharey -> share 'Price' as y
ax1.scatter(data_cleaned['year'],data_cleaned['log_price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['mileage'],data_cleaned['log_price'])
ax2.set_title('Price and Mileage')
ax3.scatter(data_cleaned['mpg'],data_cleaned['log_price'])
ax3.set_title('Price and Mpg')
ax4.scatter(data_cleaned['engineSize'],data_cleaned['log_price'])
ax4.set_title('Price and EngineSize')


plt.show()
```

```python
data_cleaned = data_cleaned.drop(['price'], axis=1)
```

### Multicollinearity

```python
# Correlation matrix
corr = data_cleaned.corr()

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)]=True
with sns.axes_style('white'):
    fig, ax = plt.subplots(figsize=(18,10))
    sns.heatmap(corr,  mask=mask, annot=True, center=0, square=True)
```

##### Moderate positive correlation between price & engineSize
##### Moderate negative correlation between year & mileage

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables = data_cleaned[['mileage','year','engineSize', 'mpg']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
vif
```

Since Year has the highest VIF, I will remove it from the model
This will drive the VIF of other variables down

```python
variables = data_cleaned[['mileage','engineSize', 'mpg']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['year'],axis=1)
```

### Create dummy variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

```python
targets = data_with_dummies['log_price']
inputs = data_with_dummies.drop(['log_price'],axis=1)
```

### Linear regression model

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

```python
inputs_scaled
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=47)
```

Create the regression

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_hat = reg.predict(X_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
```

Another useful check of our model is a residual plot
We can plot the PDF of the residuals and check for anomalies. 
In the best case scenario this plot should be normally distributed
In our case we notice that there are many negative residuals (far away from the mean)
Given the definition of the residuals (y_train - y_hat), negative values imply
that y_hat (predictions) are much higher than y_train (the targets)

```python
sns.displot(y_train-y_hat, kde=True)
plt.title('Residuals PDF');
```

```python
reg.score(X_train, y_train)
```

```python
y_hat_test = reg.predict(X_test)
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
y_test = y_test.reset_index(drop=True)
df_pf['Target'] = np.exp(y_test)
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```

```python
df_pf.describe(include='all')
```

So we have the model which predicts the price with R^2 of 0.87(And that's pretty nice result), 13,765 diffrence on average between prediction and target.

### But let's try some Ensemble methods:

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
SEED = 123
rf = RandomForestRegressor(random_state=SEED)
rf.fit(X_train, y_train)
preds = rf.predict(X_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, preds))
rs_rf = rf.score(X_test, y_test)

print(f'Random Forest RMSE is: {rmse_rf}')
print(f'Random Forest R^2 is: {rs_rf}')
```

#### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(random_state=SEED)
gb.fit(X_train, y_train)
preds = gb.predict(X_test)
rmse_gb = np.sqrt(mean_squared_error(y_test, preds))
rs_gb = gb.score(X_test, y_test)

print(f'Gradient Boosting RMSE is: {rmse_gb}')
print(f'Gradient Boosting R^2 is: {rs_gb}')
```

#### Extreme Gradient Boosting

```python
import xgboost as xgb

xgb_reg = xgb.XGBRegressor(seed=SEED)
xgb_reg.fit(X_train, y_train)
preds = xgb_reg.predict(X_test)
rmse_xgb = np.sqrt(mean_squared_error(y_test, preds))
rs_xgb = xgb_reg.score(X_test, y_test)

print(f'XGBost RMSE is: {rmse_xgb}')
print(f'XGBoost R^2 is: {rs_gb}')
```

#### Well, it seems that Random Forest has pretty much better result
Let's repeat and compare targets vs predicts

```python
df_rf = pd.DataFrame(np.exp(preds), columns=['Prediction'])
df_rf.head()
```

```python
df_rf['Target'] = np.exp(y_test)
y_test = y_test.reset_index(drop=True)
df_rf['Target'] = np.exp(y_test)
```

```python
df_rf['Residual'] = df_rf['Target'] - df_rf['Prediction']
df_rf['Difference%'] = np.absolute(df_rf['Residual']/df_rf['Target']*100)
df_rf
```

```python
df_rf.describe()
```

Well, we have nice results with Linear Regression, but because some of features in our data have non-linear shape, a linear model cannot capture the non-linear features. And in this case using some of ensemble methods is quite fine, as we can see now we have mean diffrence of 10.7% instead of 13.8% with R^2 of 0.93, instead of 0,87.
